```python
#download dataset iris 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
df = pd.read_csv(r"iris.csv")
print(df)
feature_columns = ['sepal.length', 'sepal.width', 'petal.length', 'petal.width']
X = df[feature_columns].values
y = df['variety'].values
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
y = le.fit_transform(y)
print(y)
plt.figure()
df.boxplot(by="variety", figsize=(15, 10))
plt.show()
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.metrics import confusion_matrix, accuracy_score 
classifier = KNeighborsClassifier(n_neighbors=1)
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
#confusion_matrix = confusion_matrix(y_test, y_pred)
#print(confusion_matrix)
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of the model:' + str(round(accuracy, 2)) + ' %.')

```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X=[[3,8],[4,5],[5,7],[6,3],[2,1]] #2 independent variables
Y=[-3.7,3.5,2.5,11.5,5.7]
print(X)
print(Y)
df=pd.DataFrame(X, columns=['x1','x2'] )
print(df)
df['y']=pd.Series(Y)
print(df)
#2 Train and split up 3.model(linear regression)4.intercept and coefficient 
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0,shuffle=0)
print(x_train)
print(x_test)
print(y_train)
print(y_test)
r=LinearRegression()
r.fit(x_train,y_train)
print(r.intercept_)
print(r.coef_)
#6.prediction
yp=r.predict(x_test)
print(yp)
from sklearn.metrics import *
#r square error
r_sq=r.score(x_train,y_train)
print("R square error is ",r_sq)
#5.visualisation
from mpl_toolkits.mplot3d import Axes3D
f=plt.figure(figsize=(10,10))
x=f.add_subplot(111,projection='3d')
x.scatter(df['x1'],df['x2'],df['y'])
x.set_xlabel('x1')
x.set_ylabel('x2')
x.set_zlabel('y')
```
